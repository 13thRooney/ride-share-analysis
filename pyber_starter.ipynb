```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(ride_df, city_df, 
                      how='left', on='city')

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#group merged data by city
per_city = combined_df.groupby('city')
per_city.mean()

#set average fare per city variable
avg_fare = per_city.mean()['fare']
avg_fare

#set number of rides per city variable
count_ride = per_city.count()['ride_id']
count_ride

#set total_number of drivers per city variable
count_drivers = per_city.sum()['driver_count']
count_drivers

#associate city to type
type_city = city_df.set_index('city')['type']
type_city


#create dataframe
ride_sharing_per_city_df = pd.DataFrame({
    "Number of Rides": count_ride,
    "Average Fare": avg_fare,
    "Number of Drivers": count_drivers,
    "Type of City": type_city
})
ride_sharing_per_city_df

#city type filters
urban_type = ride_sharing_per_city_df[ride_sharing_per_city_df["Type of City"] == "Urban"]
suburban_type = ride_sharing_per_city_df[ride_sharing_per_city_df["Type of City"] == "Suburban"]
rural_type = ride_sharing_per_city_df[ride_sharing_per_city_df["Type of City"] == "Rural"]


plt.suptitle('Pyber Ride Sharing Data (2016)')

# three scatter plots for each city type

plt.scatter(urban_type['Number of Rides'], urban_type['Average Fare'],
            s = urban_type['Number of Drivers']*.5,
            facecolors = "coral",
            edgecolors = "black",
            label = 'Urban',
            alpha = .75)

plt.scatter(suburban_type['Number of Rides'], suburban_type['Average Fare'],
            s = suburban_type['Number of Drivers']*.5,
            facecolors = "lightskyblue",
            edgecolors = "black",
            label = "Suburban",
            alpha = .75)

plt.scatter(rural_type['Number of Rides'], rural_type['Average Fare'],
            s = rural_type['Number of Drivers']*.5,
            facecolors = "gold",
            edgecolors = "black",
            label = "Rural",
            alpha = .75)

# Incorporate the other graph properties

plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid(b=True, which='major', axis='both')


# Create a legend
lgnd = plt.legend(frameon = True, edgecolor = 'black', title = "City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]


# Incorporate a text label regarding circle size
plt.text(42, 35, "Note : \n Circle size corrrelates with driver count per city.")


# Save Figure
plt.savefig("pyber_analysis.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

fares_per_type = combined_df.groupby('type')['fare']

sum_rural_fares = fares_per_type.sum()['Rural']
sum_suburban_fares = fares_per_type.sum()['Suburban']
sum_urban_fares = fares_per_type.sum()['Urban']

total_fares = sum_rural_fares + sum_suburban_fares + sum_urban_fares

pct_fares_rural = (sum_rural_fares / total_fares) * 100
pct_fares_suburban = (sum_suburban_fares / total_fares) * 100
pct_fares_urban = (sum_urban_fares / total_fares) * 100


# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
sizes = [pct_fares_rural, pct_fares_suburban, pct_fares_urban]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.title("% of Total Fares by City Type")

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("auto")


# Save Figure
plt.savefig("fares_by_city_type.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

rides_per_type = combined_df.groupby('type')['ride_id']
#rides_per_type.count()

count_rural_rides = rides_per_type.count()['Rural']
count_suburban_rides = rides_per_type.count()['Suburban']
count_urban_rides = rides_per_type.count()['Urban']

total_rides = count_rural_rides + count_suburban_rides + count_urban_rides

pct_rides_rural = (count_rural_rides / total_rides) * 100
pct_rides_suburban = (count_suburban_rides / total_rides) * 100
pct_rides_urban = (count_urban_rides / total_rides) * 100

# Build Pie Chart

labels = ["Rural", "Suburban", "Urban"]
sizes = [pct_rides_rural, pct_rides_suburban, pct_rides_urban]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.title("% of Total Rides by City Type")

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("auto")

# Save Figure
plt.savefig("rides_by_city_type.png")

# Show Figure
plt.show()
```
